```python
training_data=[[0,0],[1,0],[0,1],[1,1]]
target_data=[0,1,1,0]

def wrapper(lst):
    return [Vector(item) for item in lst]
x_train=wrapper(training_data)
y_target=wrapper(target_data)

model = MLP(2, [10,1]) 
print("Training Started")
learning_rate = 0.1
epochs=700

for k in range(epochs):
    for j in range(len(training_data)):
        ypred = model(x_train[j])
        loss = (ypred +(-1*y_target[j]))**2
        model.zero_grad()
        loss.backward()
        for p in model.parameters():
            for i in range(len(p.data)):
                p.data[i] += (-learning_rate) * p.grad.data[i]
    if k % 100== 0:
        print(f"Step {k}: Loss {loss.data[0]}, Pred {ypred.data[0]}")
print(f"Final: Loss {loss.data[0]:.6f}, Pred {ypred.data[0]:.4f}")
```
